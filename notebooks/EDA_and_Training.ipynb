Environment Setup

```python
# Install specialized audio and transformer libraries
# !pip install torch transformers librosa sklearn matplotlib seaborn
import torch
import librosa
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from core.wav2vec_extractor import Wav2VecFeatureExtractor
from evaluation.disjoint_tester import SpeakerDisjointSplitter
```

Exploratory Data Analysis (EDA)

```python
# Loading metadata for 78,453 segments
df = pd.read_csv('../data/metadata.csv')

# Visualizing class balance (Real vs. Synthetic)
plt.figure(figsize=(8, 5))
sns.countplot(x='label', data=df)
plt.title('Distribution of Real vs. Synthetic Audio')
plt.show()

# Verifying the 21,140 unique speakers claim
print(f"Unique Speakers in Dataset: {df['speaker_id'].nunique()}")
```

Speaker-Disjoint Splitting

```python
# Preventing data leakage by splitting at the speaker level
splitter = SpeakerDisjointSplitter()
train_df, test_df = splitter.split('../data/metadata.csv', test_size=0.2)

print(f"Training segments: {len(train_df)}")
print(f"Test segments (Speaker-Disjoint): {len(test_df)}")
```

Feature Extraction Visualization

```python
# Visualizing the IndicWav2Vec embeddings
extractor = Wav2VecFeatureExtractor()
sample_audio = train_df.iloc[0]['path']
embeddings = extractor.extract_features(sample_audio)

plt.figure(figsize=(10, 4))
plt.plot(embeddings.squeeze().numpy())
plt.title('Wav2Vec Embedding Vector (768-dimensional)')
plt.xlabel('Embedding Dimension')
plt.ylabel('Amplitude')
plt.show()
```
